```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
housing_test = pd.read_csv('housing_test.csv')

print(housing_test)
print(housing_test.shape)
```

```python
housing_train = pd.read_csv('housing_train.csv')

print(housing_train)
print(housing_train.shape)
```

```python
housing_test.isnull().any()
```

```python
housing_test = housing_test.fillna(method='ffill')
```

```python
X_test = housing_test[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
y_test = housing_test['MEDV']

X_train = housing_train[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
y_train = housing_train['MEDV']
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Ground Truth': y_test, 'Predicted': y_pred})
df.plot.scatter(x='Ground Truth', y='Predicted', color='green')

s = pd.Series(range(0, 50))
s.plot.line(color='red', linewidth=2)
```
